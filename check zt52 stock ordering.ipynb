```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import pickle
from matplotlib import pyplot as plt
import statsmodels.api as sm
from matplotlib.ticker import Formatter
import collections
from decimal import Decimal, getcontext, ROUND_HALF_UP

sh1 = pd.read_csv(r'A:\rawData\logs_20200721_zt_96_04_day_96data\mdLog_SH_20200721_0836.csv')
sz1 = pd.read_csv(r'A:\rawData\logs_20200721_zt_96_04_day_96data\mdLog_SZ_20200721_0836.csv')
sh2 = pd.read_csv(r'A:\check\20200721\zt52\mdLog_SH_20200721_0910.csv')
sz2 = pd.read_csv(r'A:\check\20200721\zt52\mdLog_SZ_20200721_0910.csv')
sh1 = sh1[(sh1['StockID'] >= 600000) & (sh1['StockID'] < 700000)]
sh2 = sh2[(sh2['StockID'] >= 600000) & (sh2['StockID'] < 700000)]
d1 = pd.concat([sh1, sz1]).sort_values(by='sequenceNo')
d2 = pd.concat([sh2, sz2]).sort_values(by='sequenceNo')
d1["time"] = d1["time"].apply(lambda x: int((x.replace(':', "")).replace(".", "")))
d2["time"] = d2["time"].apply(lambda x: int((x.replace(':', "")).replace(".", "")))
d2.loc[d2['source'].isin([20, 21]), "time"] = d2[d2['source'].isin([20, 21])]['time'] * 1000
d1 = d1[d1['StockID'].isin(d2['StockID'].unique())]
```
