```python
import pandas as pd
import numpy as np

data = pd.read_excel(r'E:\定向增发发行资料.xlsx', encoding='GBK')
data = data[data.columns[data.columns != '发行代码']]
data['发行日期'] = data['发行日期'].astype(str)
data['定增股份上市日'] = data['定增股份上市日'].astype(str)
data['限售解禁日'] = data['限售解禁日'].astype(str)
add1 = pd.read_excel(r'E:\hs300.xlsx', encoding='GBK')
data = pd.merge(data, add1[['交易日期', '收盘价']], left_on='发行日期', right_on='交易日期', how='left')
data = data[data.columns[data.columns!='交易日期']]
data = data.rename(columns={'收盘价':'沪深300_发行日'})
data = pd.merge(data, add1[['交易日期', '收盘价']], left_on='定增股份上市日', right_on='交易日期', how='left')
data = data[data.columns[data.columns!='交易日期']]
data = data.rename(columns={'收盘价':'沪深300_股份上市日'})
```

```python
import numpy as np
import pandas as pd
import glob

def clean(x):
    cl_ean = x["ListDays"] != 0
    x = x[cl_ean]
    cl_ean1 = x["trade_stats"] == 1
    x = x[cl_ean1]
    return x.iloc[:,[0,1,2,3,5,6,7,8,9,10,11,12,13]]

path = r'E:\stockDaily\data'
all_files = glob.glob(path + "\*.csv")
dd = clean(pd.read_csv(all_files[0], encoding="GBK"))
for i in range(1, len(all_files)):
    dn = clean(pd.read_csv(all_files[i], encoding="GBK"))
    dd = pd.concat([dd, dn], axis = 0, ignore_index = True)
dd = dd.sort_values(by=["Symbol", "Date"])
dd
```

```python
dd['流通市值'] = dd['Marketshares'] * dd['close']
dd['代码'] = dd['Symbol'].str[2:] + '.' + dd['Symbol'].str[:2]
dd['发行日期'] = dd['Date'].astype(str).str[:4] + '-' + dd['Date'].astype(str).str[4:6] + '-' + dd['Date'].astype(str).str[6:8]
dd.loc[dd['代码'] == '000022.SZ', '代码'] = '001872.SZ'
dd.loc[dd['代码'] == '000043.SZ', '代码'] = '001914.SZ'
dd.loc[dd['代码'] == '601313.SH', '代码'] = '601360.SH'
```

```python
data = pd.merge(data, dd[['代码', '发行日期', '流通市值']], on=['代码', '发行日期'], how='left')
data[data['流通市值'].isnull()]
```

```python
add = pd.read_excel(r'E:\指数行情序列.xlsx')
```

```python
df = []
for i in add.columns[1:]:
    cl = ['时间'] + [i]
    ad = add[cl]
    ad = ad.rename(columns={i:'收盘价'})
    ad['指数'] = 'Wind'+i[:-2]
    df += [ad]
df = pd.concat(df)
df['时间'] = df['时间'].astype(str)
```

```python
data = pd.merge(data, df, left_on=['定增股份上市日', 'Wind行业'], right_on=['时间', '指数'], how='left')
data = data[data.columns[(data.columns != '时间') & (data.columns != '指数')]]
data = data.rename(columns={'收盘价':'指数_定增股份上市日'})
```

```python
data = pd.merge(data, df, left_on=['限售解禁日', 'Wind行业'], right_on=['时间', '指数'], how='left')
data = data[data.columns[(data.columns != '时间') & (data.columns != '指数')]]
data = data.rename(columns={'收盘价':'指数_限售解禁日'})
```

```python
data.to_csv(r'E:\定向增发发行资料.csv', encoding='GBK', index=False)
```
