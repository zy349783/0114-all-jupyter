```python
import pandas as pd
import random
import numpy as np
import glob
import os
from unrar import rarfile
import py7zr
import pickle
import datetime
import time
pd.set_option("max_columns", 200)


year = "2018"
startDate = '20180102'
endDate = '20180629'
bad = []
less = []
readPath = 'K:\\data\\' + year + '\\***\\***'
dataPathLs = np.array(glob.glob(readPath))
dateLs = np.array([os.path.basename(i).split('_')[0] for i in dataPathLs])
dataPathLs = dataPathLs[(dateLs >= startDate) & (dateLs <= endDate)]
date_list = pd.read_csv("G:\\trading_days.csv")
for data in dataPathLs:
    if len(np.array(glob.glob(data + '\\SH\\***'))) == 0:
        if int(os.path.basename(data)) not in date_list["Date"].values:
            continue
        else:
            print(os.path.basename(data) + " less data!!!!!!!!!!!!!!!!!")
            less.append(data)
            continue
    
    startTm = datetime.datetime.now()
    try:
        a = py7zr.SevenZipFile(data + '\\SH\\snapshot.7z','r',filters=None)
    except:
        print("Bad unzip here!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!")
        print(data + '\\SH\\snapshot.7z')
        bad.append(data + '\\SH\\snapshot.7z')
        continue
    date = os.path.basename(data)
    path = 'E:\\KR\\' + year + '\\SH'
    path1 = path + '\\' + date
    os.mkdir(path1)
    a.extractall(path = path1)
    a.close()
    
    try:
        a = py7zr.SevenZipFile(data + '\\SZ\\snapshot.7z','r',filters=None)
    except:
        print("Bad unzip here!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!")
        print(data + '\\SH\\snapshot.7z')
        bad.append(data + '\\SH\\snapshot.7z')
        continue
    date = os.path.basename(data)
    path = 'E:\\KR\\' + year + '\\SZ'
    path1 = path + '\\' + date
    os.mkdir(path1)
    a.extractall(path = path1)
    a.close()
    print(datetime.datetime.now() - startTm)
    print(date + 'finished')
    
print(len(dataPathLs))
print(len(np.array(glob.glob('E:\\KR\\' + year + '\\SZ\\***'))))
print(len(np.array(glob.glob('E:\\KR\\' + year + '\\SH\\***'))))
print(bad)
print(less)
```
