```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import pickle
from matplotlib import pyplot as plt
import statsmodels.api as sm
order = pd.read_table(r"D:\work\project 13 lv2 data check\SZL2_ORDER_20150505\SZL2_ORDER_20150505_AM.txt", header=None)
order = pd.concat([order, pd.read_table(r"D:\work\project 13 lv2 data check\SZL2_ORDER_20150505\SZL2_ORDER_20150505_PM.txt", header=None)])
cols = pd.read_csv(r"C:\Users\win\Desktop\kk.csv", encoding="utf-8", header=None)
order.columns = cols.iloc[:, 0].values
order = order.sort_values(by="RECNO")
order = order[((order["SECURITYID"] >= 1) & (order["SECURITYID"] < 4000)) |((order["SECURITYID"] >= 300000) & (order["SECURITYID"] < 310000))]
order.head()
```

```python
# 1.Suppose it is similiar to immediate order when "price"==0, buy orders
d1 = order[(order["ORDERKIND"] == "Y ") & (order["PRICE"] == 0) & (order["FUNCTIONCODE"] != 'C ')]
d1e = pd.merge(trade[["BUYORDERRECNO", "SECURITYID", "PRICE", "TRADEQTY", "ORDERKIND", "FUNCTIONCODE", 'TRADETIME']], 
         d1[d1["FUNCTIONCODE"] == "B "][["SECURITYID", "ORDERQTY", "ORDERENTRYTIME", "RECNO"]], left_on=["SECURITYID","BUYORDERRECNO"],
        right_on=["SECURITYID","RECNO"], how="right")
print(d1e[d1e["TRADEQTY"].isnull()])
k = d1e.groupby(["BUYORDERRECNO", "SECURITYID"])["TRADEQTY"].size().reset_index()
k1 = k[k["TRADEQTY"] == 1]
k2 =  k[k["TRADEQTY"] != 1]
display(k1.head())
display(k2.head())
k1 = pd.merge(k1, d1e, on=["BUYORDERRECNO", "SECURITYID"])
k2 = pd.merge(k2, d1e, on=["BUYORDERRECNO", "SECURITYID"])
display(k1[k1["TRADEQTY_y"] != k1["ORDERQTY"]])
display(order[(order["ORDERKIND"] == "Y ") & (order["SECURITYID"] == 2622) & (order["RECNO"] == 1700039)])
display(trade[(trade["SECURITYID"] == 2622) & (trade["BUYORDERRECNO"] == 1700039)])
k2["TotalTrade"] = k2.groupby(["BUYORDERRECNO", "SECURITYID"])["TRADEQTY_y"].transform("sum")
k22 =k2.groupby(["BUYORDERRECNO", "SECURITYID"])["TotalTrade", "ORDERQTY","TRADETIME","ORDERENTRYTIME"].first().reset_index()
display(k22[k22["TotalTrade"] != k22["ORDERQTY"]])
display(order[(order["ORDERKIND"] == "Y ") & (order["SECURITYID"] == 2622) & (order["RECNO"] == 1703362)])
display(trade[(trade["SECURITYID"] == 2622) & (trade["BUYORDERRECNO"] == 1703362)])
```

```python
# 2.Suppose it is similiar to immediate order when "price"==0, sell orders
d1 = order[(order["ORDERKIND"] == "Y ") & (order["PRICE"] == 0) & (order["FUNCTIONCODE"] != 'C ')]
d1e = pd.merge(trade[["SELLORDERRECNO", "SECURITYID", "PRICE", "TRADEQTY", "ORDERKIND", "FUNCTIONCODE", 'TRADETIME']], 
         d1[d1["FUNCTIONCODE"] == "S "][["SECURITYID", "ORDERQTY", "ORDERENTRYTIME", "RECNO"]], left_on=["SECURITYID","SELLORDERRECNO"],
        right_on=["SECURITYID","RECNO"], how="right")
print(d1e[d1e["TRADEQTY"].isnull()])
k = d1e.groupby(["SELLORDERRECNO", "SECURITYID"])["TRADEQTY"].size().reset_index()
k1 = k[k["TRADEQTY"] == 1]
k2 =  k[k["TRADEQTY"] != 1]
display(k1.head())
display(k2.head())
k1 = pd.merge(k1, d1e, on=["SELLORDERRECNO", "SECURITYID"])
k2 = pd.merge(k2, d1e, on=["SELLORDERRECNO", "SECURITYID"])
display(k1[k1["TRADEQTY_y"] != k1["ORDERQTY"]])
# display(order[(order["ORDERKIND"] == "Y ") & (order["SECURITYID"] == 2622) & (order["RECNO"] == 1700039)])
# display(trade[(trade["SECURITYID"] == 2622) & (trade["SELLORDERRECNO"] == 1700039)])
k2["TotalTrade"] = k2.groupby(["SELLORDERRECNO", "SECURITYID"])["TRADEQTY_y"].transform("sum")
k22 =k2.groupby(["SELLORDERRECNO", "SECURITYID"])["TotalTrade", "ORDERQTY","TRADETIME","ORDERENTRYTIME"].first().reset_index()
display(k22[k22["TotalTrade"] != k22["ORDERQTY"]])
# display(order[(order["ORDERKIND"] == "Y ") & (order["SECURITYID"] == 2622) & (order["RECNO"] == 1703362)])
# display(trade[(trade["SECURITYID"] == 2622) & (trade["SELLORDERRECNO"] == 1703362)])
k3 = k2[k2["FUNCTIONCODE"] == 'C '][["SELLORDERRECNO", "SECURITYID"]]
k3 = pd.merge(k3, d1e, on=["SELLORDERRECNO", "SECURITYID"])
display(k3.head(50))
```
