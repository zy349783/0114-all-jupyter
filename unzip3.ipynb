```python
import pandas as pd
import random
import numpy as np
import glob
import os
from unrar import rarfile
import py7zr
import pickle
import datetime
import time
pd.set_option("max_columns", 200)


year = "2019"
startDate = '20190521'
endDate = '20191001'
bad = []
less = []
l = []
readPath = 'K:\\data\\' + year + '\\***\\***'
dataPathLs = np.array(glob.glob(readPath))
dateLs = np.array([os.path.basename(i).split('_')[0] for i in dataPathLs])
dataPathLs = dataPathLs[(dateLs >= startDate) & (dateLs <= endDate)]
date_list = pd.read_csv("G:\\trading_days.csv")
for data in dataPathLs:
    if len(np.array(glob.glob(data + '\\SH\\***'))) == 0:
        if int(os.path.basename(data)) not in date_list["Date"].values:
            continue
        else:
            print(os.path.basename(data) + " less data!!!!!!!!!!!!!!!!!")
            less.append(data)
            continue
    
    startTm = datetime.datetime.now()
    date = os.path.basename(data)
    exe_path = 'G:\\7z1900-extra\\7za.exe'
    rar_path = data + '\\SH\\snapshot.7z'
    path = 'G:\\KR\\' + year + '\\SH'
    path1 = path + '\\' + date
    os.mkdir(path1)
    un_path = path1
    cmd = '{} x {} -o{} -aos -r'.format(exe_path, rar_path, un_path)
    os.system(cmd)

    date = os.path.basename(data)
    exe_path = 'G:\\7z1900-extra\\7za.exe'
    rar_path = data + '\\SZ\\snapshot.7z'
    path = 'G:\\KR\\' + year + '\\SZ'
    path1 = path + '\\' + date
    un_path = path1
    cmd = '{} x {} -o{} -aos -r'.format(exe_path, rar_path, un_path)
    os.system(cmd)
    
    date = os.path.basename(data)
    exe_path = 'G:\\7z1900-extra\\7za.exe'
    rar_path = data + '\\SZ\\order.7z'
    path = 'G:\\KR\\' + year + '\\SZ'
    path1 = path + '\\' + date
    un_path = path1
    cmd = '{} x {} -o{} -aos -r'.format(exe_path, rar_path, un_path)
    os.system(cmd)
    
    date = os.path.basename(data)
    exe_path = 'G:\\7z1900-extra\\7za.exe'
    rar_path = data + '\\SZ\\tick.7z'
    path = 'G:\\KR\\' + year + '\\SZ'
    path1 = path + '\\' + date
    un_path = path1
    cmd = '{} x {} -o{} -aos -r'.format(exe_path, rar_path, un_path)
    os.system(cmd)
    
    l.append(data)
    print(datetime.datetime.now() - startTm)
    print(date + 'finished')
    
print(len(l))
print(bad)
print(less)
```
